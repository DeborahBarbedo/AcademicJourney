# Bibliotecas

```python
import os
import pandas as pd

import math

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

# Importação dos dados

```python
d_path = os.getcwd()
df = pd.read_csv(d_path+"\\estacoes.csv"
                     ,sep = ','
                     ,decimal='.'
                      )
```

```python
df
```

```python
quantidade_estacoes = len(df["id_estacao"])
latitude = df["latitude"]
longitude = df["long"]
```

```python
M = np.zeros((quantidade_estacoes,quantidade_estacoes))
lat, lon = [min(latitude),min(longitude)]
valor_andar_lat = (max(latitude)-min(latitude))/quantidade_estacoes
valor_andar_lon = (max(longitude)-min(longitude))/quantidade_estacoes
latitude = df["latitude"]
longitude = df["long"]
id_medidor = df["id_estacao"]
for y in range(quantidade_estacoes): #Percorrendo longitudes
    for x in range(quantidade_estacoes): #Percorrendo latitudes
        dmin = math.hypot(lat-valor_andar_lat, lon-valor_andar_lon) #Norma euclidiana, distância mínima entre máximo de x e máximo de y
        j = -1
        for i in range(quantidade_estacoes):
            d = math.hypot(latitude[i]-lat-(x*valor_andar_lat), longitude[i]-lon-(y*valor_andar_lon)) #Norma euclidiana, distância mínima entre a diferença do valor a cada ponto do loop
            if d < dmin: # Se distancia for menor que a distancia minima estipulada
                dmin = d # distancia minima recebe novo valor
                j = id_medidor[i] #  (Colocar ID_medidor)
        M[x][y] = j
```

```python
M
```

```python
data_frame = pd.DataFrame(M)
```

```python
data_frame
```

```python
lat_df = []
lon_df = []
lat, lon = [min(latitude),min(longitude)]
valor_andar_lat = (max(latitude)-min(latitude))/quantidade_estacoes
valor_andar_lon = (max(longitude)-min(longitude))/quantidade_estacoes
for i in range(quantidade_estacoes):
    lat_df.append(lat+(i*valor_andar_lat))
    lon_df.append(lon+(i*valor_andar_lon))
```
