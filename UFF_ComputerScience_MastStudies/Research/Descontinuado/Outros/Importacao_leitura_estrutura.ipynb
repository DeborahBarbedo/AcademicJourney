# Bibliotecas

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
import matplotlib.ticker as ticker
```

# Leitura dos arquivos

```python
# Caminho do documento atual
d_path = os.getcwd()

# Nome do arquivo
arquivo_chuvas = "\\chuvas_2020.csv"
arquivo_estacao = "\\data.csv"
arquivo_deslizamentos = "\\deslizamento-geocoded.xlsx"

# Importação dos dados

df_chuvas = pd.read_csv(d_path+arquivo_chuvas
                     ,sep = ','
                     ,decimal='.'
                      )
df_estacao = pd.read_csv(d_path+arquivo_estacao
                     ,sep = ';'
                     ,decimal=','
                      )

df_deslizamentos = pd.read_excel(d_path+arquivo_deslizamentos
                                 ,engine='openpyxl'
                                )
```

# Criação da base de estações

```python
df_estacao = df_estacao.groupby(['codEstacao', 'nomeEstacao', 'latitude', 'longitude']).count().reset_index()[['codEstacao', 'latitude', 'longitude']]
```
